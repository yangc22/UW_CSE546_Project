```python
import pandas as pd
```

```python
def change_name(folder, file_name, file_format, season):
    suffix = 'new'
    df = pd.read_csv("{}/{}_{}.{}".format(folder, file_name, suffix, file_format))
    df.to_csv("{}/{}_{}.{}".format(folder, file_name, season, file_format), encoding='utf-8', index=False)
```

```python
folder = '/Users/chaoyiyang/Documents/Course/CSE546-Machine Learning-18Au/Project/UW_CSE546_Project/ELO Rating/training_data'
files = ['TeamPerGameStats', 'OpponentPerGameStats', 'MiscellaneousStats']
file_format = 'csv'
season = '1819'
for file_name in files:
    change_name(folder, file_name, file_format, season)
```

```python
def append_different_seasons(folder, file_name, file_format, season1, season2):
    df1 = pd.read_csv("{}/{}_{}.{}".format(folder, file_name, season1, file_format))
    df2 = pd.read_csv("{}/{}_{}.{}".format(folder, file_name, season2, file_format))
    print(df1)
    print(df2)
    df = df1.append(df2, ignore_index=True)
    print(df)
```
