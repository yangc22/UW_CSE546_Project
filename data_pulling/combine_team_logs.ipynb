```python
import pandas as pd
```

```python
def combine_season_games(season1, season2):
    stats_folder = ''
    file_name = 'season_games'
    file_type = '.xlsx'
    df1 = pd.read_excel("{}{}_{}{}".format(stats_folder, season1, file_name, file_type))
    df1 = df1[df1['schedule.playedStatus'] == 'COMPLETED']
    df2 = pd.read_excel("{}{}_{}{}".format(stats_folder, season2, file_name, file_type))
    # df2 = df2[df2['schedule.playedStatus'] == 'COMPLETED']
    df = df1.append(df2, ignore_index=True)
    dropped = ['schedule.attendance', 'schedule.broadcasters', 'schedule.delayedOrPostponedReason', 'schedule.endedTime', 'schedule.officials', 'schedule.originalStartTime', 'schedule.weather', 'score.currentIntermission', 'score.currentQuarter', 'score.currentQuarterSecondsRemaining', 'score.quarters']
    df = df[df.columns.drop(dropped)]
    output_folder = ''
    output_format = 'csv'
    # save the df
    df.to_csv("{}{}_{}.{}".format(output_folder, 'combined', file_name, output_format), encoding='utf-8', index=False)
    # return the df for debugging
    # return df
```

```python
def combine_stats(team_name):
    stats_folder = 'team_game_logs_'
    file_name = 'team_game_log'
    season1 = '1718'
    season2 = '1819'
    file_type = '.xlsx'
    df1 = pd.read_excel("{}{}/{}_{}{}".format(stats_folder, season1, file_name, team_name, file_type))
    df2 = pd.read_excel("{}{}/{}_{}{}".format(stats_folder, season2, file_name, team_name, file_type))
    df = df1.append(df2, ignore_index=True)
    df = df[df.columns.drop(list(df.filter(regex='PerGame')))]
    output_folder = 'team_game_logs_combined'
    output_format = 'csv'
    l_id = df['team.id'].tolist()
    print(team_name, set(l_id))
    df.to_csv("{}/{}_{}.{}".format(output_folder, file_name, team_name, output_format), encoding='utf-8', index=False)
    # return df
```

```python
df = combine_stats('ATL')
```
